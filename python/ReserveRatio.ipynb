```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the supply range and multiplier for the price functions
supply_range = np.linspace(0, 10, 10)  # Start at 1 to avoid division by zero for RR=1
m = 20  # Set the multiplier to 1 for simplicity, this can be scaled as needed

# Define the price functions for different reserve ratios
def price_flat(supply):
    return np.full_like(supply, m)  # Returns a constant price

def price_linear(supply):
    return m * supply

def price_polynomial(supply):
    return m * supply**9

def price_root(supply):
    return m * supply**(1/9)

# Plot the bonding curves for different reserve ratios
plt.figure(figsize=(12, 8))

plt.plot(supply_range, price_flat(supply_range), label='RR = 1 (flat)')
plt.plot(supply_range, price_linear(supply_range), label='RR = 1/2 (linear)')
plt.plot(supply_range, price_polynomial(supply_range), label='RR = 1/10 (polynomial, n=9)')
plt.plot(supply_range, price_root(supply_range), label='RR = 9/10 (root, n=1/9)')

plt.title('Bonding Curve Prices for Different Reserve Ratios')
plt.xlabel('Continuous Token Supply')
plt.ylabel('Price (f(x))')
plt.legend()
plt.grid(True)
plt.show()
```
