```python
import matplotlib.pyplot as plt
import numpy as np

# Define a range of supply values
supply = np.linspace(0, 10, 100)  # Supply values from 0 to 10
p = supply ** 2  # Compute p as the square of supply values

# Plotting the graph
plt.plot(supply, p)
plt.title("Supply Squared Graph")
plt.xlabel("Supply")
plt.ylabel("p (Supply Squared)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set parameters
reserve_balance = 10000  # Assume a reserve balance
reserve_ratio = 0.5  # Reserve ratio (50%)

# Create a range of total supply values from 1 to 1000
total_supply_values = np.linspace(1, 1000, 1000)  # Start from 1 to avoid division by zero

# Calculate continuous token price (cp) for each supply value
# This assumes a bonding curve where price increases with supply and a fixed reserve ratio
cp_values = (reserve_balance / (reserve_ratio * total_supply_values))

# Plot the graph
plt.plot(total_supply_values, cp_values)
plt.title('Continuous Token Price as a function of Total Supply with a 50% Reserve Ratio')
plt.xlabel('Total Supply of Tokens')
plt.ylabel('Price per Token (Ether)')
plt.grid(True)
plt.show()
```
